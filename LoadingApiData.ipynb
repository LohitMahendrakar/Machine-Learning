```python
import http.client

conn = http.client.HTTPSConnection("youtube138.p.rapidapi.com")

payload = "{\"id\":\"UCJ5v_MCY6GNUBTO8-D3XoAg\",\"filter\":\"videos_latest\",\"cursor\":\"\",\"hl\":\"en\",\"gl\":\"US\"}"

headers = {
    'x-rapidapi-key': "786da3c4d9mshefb5415899de265p11ca59jsne468c85bfcc8",
    'x-rapidapi-host': "youtube138.p.rapidapi.com",
    'Content-Type': "application/json"
}

conn.request("POST", "/channel/videos/", payload, headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```

```python
import json

json_data = json.loads(data.decode("utf-8"))
```

```python
rows = []

for item in json_data["contents"]:
    if item["type"] == "video":
        video = item["video"]
        
        rows.append({
            "video_id": video.get("videoId"),
            "title": video.get("title"),
            "views": video.get("stats", {}).get("views"),
            "length_seconds": video.get("lengthSeconds"),
            "published": video.get("publishedTimeText")
        })
```

```python
import pandas as pd

df = pd.DataFrame(rows)
df
```
